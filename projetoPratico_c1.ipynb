```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('./MICRODADOS.CSV', sep=';', encoding='latin1')
df
```

```python
df.columns
```

```python
mortes = df.query('(Classificacao=="Confirmados") &  (Evolucao=="Óbito pelo COVID-19") ').drop_duplicates()
fato_mortes = mortes.groupby(['Municipio', 'Sexo', 'RacaCor']).agg({'Evolucao': np.count_nonzero}).reset_index()
fato_mortes.rename(columns={'Evolucao': 'Mortes'}, inplace=True)
fato_mortes
```

```python
dim_datas = mortes[['DataNotificacao', 'DataCadastro', 'DataDiagnostico', 'DataColeta_RT_PCR', 'DataEncerramento', 'DataObito']]
dim_datas.sample(5)
```

```python
dim_escolaridade = mortes[['Sexo','RacaCor', 'Escolaridade', "Municipio"]]
dim_escolaridade
```

```python
dim_comorbidades = mortes[['ComorbidadePulmao', 'ComorbidadeCardio',
       'ComorbidadeRenal', 'ComorbidadeDiabetes', 'ComorbidadeTabagismo',
       'ComorbidadeObesidade']]
dim_comorbidades
```

```python
base = mortes[['FicouInternado', 'ProfissionalSaude', 'PossuiDeficiencia']]
dim_profissional_saude = base.query('ProfissionalSaude == "Sim" ')
dim_profissional_saude.sample(5)
```

```python
# realizando a conexão com o banco de dados postgres localmente, via docker, e acessando o database projeto_bi
engine = create_engine('postgresql://root:root@localhost:5432/projeto_bi')
```

```python
# df.to_sql('nome_da_tabela', con=engine, if_exists='replace', index=False)
fato_mortes.to_sql('fato_mortes', con=engine, if_exists='replace', index=False)
dim_datas.to_sql('dim_datas', con=engine, if_exists='replace', index=False)
dim_escolaridade.to_sql('dim_escolaridade', con=engine, if_exists='replace', index=False)
dim_comorbidades.to_sql('dim_comorbidades', con=engine, if_exists='replace', index=False)
dim_profissional_saude.to_sql('dim_profissional_saude', con=engine, if_exists='replace', index=False)
```
